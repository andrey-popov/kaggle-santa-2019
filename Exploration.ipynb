```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
families_data = pd.read_csv('family_data.csv')
families_data.head()
```

```python
preference_occupancy = np.zeros((10, 100), dtype=np.int32)
for _, row in families_data.iterrows():
    size = row['n_people']
    for choice in range(10):
        day = row['choice_{}'.format(choice)]
        preference_occupancy[choice, day - 1] += size
```

```python
fig = plt.figure(figsize=(20, 30))
for choice in range(10):
    axes = fig.add_subplot(5, 2, choice + 1)
    axes.bar(range(1, 101), preference_occupancy[choice])
    axes.set_title('Choice #{}'.format(choice))
    axes.set_xlim(0., 101.)
```
